```python
february_data = df.loc['2020-02-01':'2020-02-29']
february_data
```

```python
sns.lineplot(february_data["Adj Close"])
plt.xlabel("")
plt.ylabel("Adjusted Close Price")
plt.title("Alibaba Adjusted Close Price in February 2020")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(february_data["Adj Close"], color="red", linestyle="--")
plt.xlabel("")
plt.ylabel("Adjusted Close Price")
plt.title("Alibaba Adjusted Close Price in February 2020")
plt.xticks(rotation=45)
plt.show()
```

```python
january_data = df.loc['2020-01-01':'2020-01-31']
january_data.head()

```

```python
sns.lineplot(x=january_data.index.day, y=january_data["Adj Close"], color="black")
sns.lineplot(x=february_data.index.day, y=february_data["Adj Close"], color="red", linestyle="--")
plt.xlabel("")
plt.ylabel("Adjusted Close Price")
plt.title("Alibaba Adjusted Close Price in February 2020")
plt.xticks(rotation=45)
plt.show()
```

```python
march_data = df.loc['2020-03-01':'2020-03-31']
march_data
```

```python
sns.lineplot(x=january_data.index.day, y=january_data["Adj Close"], color="black")
sns.lineplot(x=february_data.index.day, y=february_data["Adj Close"], color="red", linestyle="--")
sns.lineplot(x=march_data.index.day, y=march_data["Adj Close"], color="gray")
plt.xlabel("")
plt.ylabel("Adjusted Close Price")
plt.title("Alibaba Adjusted Close Price Jan-Mar 2020")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(x=january_data.index.day, y=january_data["Adj Close"], color="black", label="January")
sns.lineplot(x=february_data.index.day, y=february_data["Adj Close"], color="red", linestyle="--", label="February")
sns.lineplot(x=march_data.index.day, y=march_data["Adj Close"], color="gray", label="March")
plt.xlabel("")
plt.ylabel("Adjusted Close Price")
plt.title("Alibaba Adjusted Close Price Jan-Mar 2020")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
ax = sns.lineplot(data=df['Adj Close'], color="orange")
plt.title("Alibaba Adjusted Close Price (Full Range)")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()


ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.show()
```

```python
ax = sns.lineplot(data=df['Volume'], color="orange")
plt.title("Alibaba Volume (Full Range)")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()


ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.show()
```

```python
# calculate moving averages
Moving_Avg_14 = df['Volume'].rolling(window=14).mean()
Moving_Avg_14
```

```python
plt.figure(figsize=(12,6))  

ax = sns.lineplot(data=df['Volume'], color="orange", alpha=0.5, linestyle=":", label="Daily Volume")
sns.lineplot(data=Moving_Avg_14, color="blue", label="14-Day Moving Average")
plt.title("Alibaba Volume (Full Range)")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()


ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.show()
```

```python
Moving_Avg_90 = df['Volume'].rolling(window=90).mean()
Moving_Avg_90
```

```python
plt.figure(figsize=(12,6))  

ax = sns.lineplot(data=df['Volume'], color="orange", alpha=0.5, linestyle=":", label="Daily Volume")
sns.lineplot(data=Moving_Avg_90, color="blue", label="90-Day Moving Average")
plt.title("Alibaba Volume (Full Range)")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()


ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.show()
```

```python
# Investigate adjusted close price
pct_change_ali = df['Adj Close'].pct_change()
pct_change_ali.head()
```

```python
pct_change_ali*100
```

```python
plt.figure(figsize=(12,6))  

sns.lineplot(data=pct_change_ali*100, color="blue")
plt.title("Alibaba Moving Average -  pct change")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()


ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=pct_change_ali*100, color="blue")
plt.axhline(y=-5, color="red", linestyle="--", linewidth=1, label="-5%")
plt.axhline(y=0, color="black", linewidth=1, label="0%")  # -0.05  100 = -5%
plt.title("Alibaba Moving Average - pct change")
plt.xlabel("", fontsize=14)
plt.ylabel("Adjusted Close Price", fontsize=14)
plt.grid()

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("'%y"))
plt.legend()
plt.show()
```
