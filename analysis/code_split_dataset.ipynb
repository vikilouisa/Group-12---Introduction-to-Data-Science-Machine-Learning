```python
import pandas as pd
from pathlib import Path

# Load merged file
df = pd.read_csv("merged_data_updated.csv", parse_dates=["date"])

# Date ranges
train_start, train_end = "2013-07-01", "2017-07-31"
val_start,   val_end   = "2017-08-01", "2018-07-31"
test_start,  test_end  = "2018-08-01", "2019-07-31"

# Splits
train_df = df[(df["date"] >= train_start) & (df["date"] <= train_end)].copy()
val_df   = df[(df["date"] >= val_start)   & (df["date"] <= val_end)].copy()
test_df  = df[(df["date"] >= test_start)  & (df["date"] <= test_end)].copy()

print("Train rows:", len(train_df))
print("Validation rows:", len(val_df))
print("Test rows:", len(test_df))

print("\nFirst rows Testset:")
print(test_df.head())

# ---------------------------------------------------------
# Save splits to your existing CSV files
# ---------------------------------------------------------
train_df.to_csv("train_split_merged_data_updated.csv", index=False)
val_df.to_csv("val_split_merged_data_updated.csv", index=False)
test_df.to_csv("test_split_merged_data_updated.csv", index=False)

print("\nUpdated split files saved:")
print("train_split_merged_data_updated.csv")
print("val_split_merged_data_updated.csv")
print("test_split_merged_data_updated.csv")
```
