```python
import pandas as pd

nn = pd.read_csv("submission_neural_net.csv")
lr = pd.read_csv("submission_linear_regression_byGulfem_2.csv")

# IMPORTANT: use the SAME filtered test file you used for NN
test = pd.read_csv("test_split_merged_expanded_data_filtered.csv")[["id", "warengruppe"]]
```

```python
assert set(nn.id) == set(lr.id)
```

```python
df = test.merge(nn, on="id", how="left", suffixes=("", "_nn"))
df = df.merge(lr, on="id", how="left", suffixes=("", "_lr"))

# rename for clarity
df = df.rename(columns={
    "umsatz_Prediction": "pred_nn",
    "umsatz_Prediction_lr": "pred_lr"
})
```

```python
use_nn = df["warengruppe"].isin([1, 2, 3])

df["umsatz_Prediction"] = df["pred_lr"]        # default = LR
df.loc[use_nn, "umsatz_Prediction"] = df.loc[use_nn, "pred_nn"]
```

```python
print(df["umsatz_Prediction"].isna().sum())  # should be 0
print(df.shape)                              # should match Kaggle rows
```

```python
hybrid = df[["id", "umsatz_Prediction"]].sort_values("id")
hybrid.to_csv("submission_hybrid_nn_lr.csv", index=False)

print("Saved: submission_hybrid_nn_lr.csv")
hybrid.head()
```
